```python
import re #for regular expression
import pandas as pd
```

```python
# Initially txt file is readed in binary format(01010), so to convert it into string we use encoding='utf-8'.
# r denotes reading format
f=open("WhatsApp Chat with Soft eng activity (1).txt","r",encoding='utf-8')
```

```python
data=f.read()
```

```python
print(data)
```

```python
# pattern is a regular expression
# \ is used to divide the digits
# / is directly taken from the string above
# \s denotes giving space 
# , is directly taken from the string above
# - is directly taken from the string above
pattern='\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
# converting string into list
messages=re.split(pattern, data)[1:]
len(messages)
```

```python
messages
```

```python
# findall is used to find all the dates in list format
dates=re.findall(pattern, data)
dates
```

```python
# converting list into dataframe
df=pd.DataFrame({'user_message': messages, 'message_date': dates})
#convert message_data type
df['message_date']=pd.to_datetime(df['message_date'], format = "%m/%d/%y, %H:%M - ")
df.rename(columns={'message_date':'date'}, inplace=True)
df.head()
```

```python
df.shape
```

```python
users=[]
messages=[]
for message in df['user_message']:
    entry=re.split('([\w\W]+?):\s', message)
    
    if entry[1:]: #user name
        users.append(entry[1])
        messages.append(entry[0])
    else:
        users.append('group_notification')
        messages.append(entry[0])
        
df['user']=users
df['message']=messages
df.drop(columns=['user_message'], inplace=True)

df.head()
```
